```python
import numpy as np

# Task 1: Moving Average (विंडो साइज 3 का मूविंग एवरेज)
prices = np.array([10, 20, 30, 40, 50, 60])
moving_avg = np.convolve(prices, np.ones(3)/3, mode='valid')
print("Moving Average:", moving_avg)

# Task 2: Boolean Grid Search (50 से बड़ा और 5 से विभाज्य कोई नंबर है या नहीं)
a = np.array([[54, 54, 11, 42],
              [13, 13, 96, 20],
              [39,  2, 96, 24],
              [83, 75, 31, 35]])
has_match = np.any((a > 50) & (a % 5 == 0))
print("\nIs there any match?:", has_match)

# Task 3: Arg-Sort Ranker (छोटे से बड़े मार्क्स के मूल इंडेक्स का ऑर्डर)
marks = np.array([75, 92, 40, 85, 60])
sorted_indices = np.argsort(marks)
print("\nSorted Indices:", sorted_indices)

# Task 4: One-Hot Encoder (कैटेगरी लेबल्स को बाइनरी ग्रिड में बदलना)
labels = np.array([0, 2, 1, 3])
one_hot = (labels[:, None] == np.arange(4)).astype(int)
print("\nOne-Hot Encoded Matrix:\n", one_hot)

# Task 5: Matrix Cross Product (टॉर्क वेक्टर r x F निकालना)
r = np.array([1, 0, -2])
F = np.array([2, 3, 4])
torque = np.cross(r, F)
print("\nTorque Vector:", torque)
```
